```python
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler

from sklearn.ensemble import RandomForestClassifier

import joblib
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
newdf = df.drop(columns= ['koi_steff','koi_slogg','koi_srad', 'koi_srad_err2','koi_slogg_err1','koi_tce_plnt_num'], axis=1)

newdf.head()
```

```python
X = newdf.drop("koi_disposition", axis=1)

```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = newdf["koi_disposition"]

# Split data into training and testing groups
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
model = RandomForestClassifier(n_estimators=100)
model = model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
model = RandomForestClassifier(random_state=42)

param_grid = {
   'n_estimators':[100, 200],
    'max_features':['auto', 'sqrt'],
    'max_depth': [14, 15, None]
}
```

```python
# Train the model with GridSearch
grid = GridSearchCV(model, param_grid, cv=5, verbose=0)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
print(f"Training Data Score: {grid.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {grid.score(X_test_scaled, y_test)}")
```

```python
predictions = grid.predict(X_test_scaled)
classifications = y_test.unique().tolist()

prediction_actual = {
    'Actual': y_test,
    'Prediction': predictions
}

prediction_df = pd.DataFrame(prediction_actual)
PA_df = prediction_df.set_index('Actual').reset_index()
prediction_df.head(15)
```
